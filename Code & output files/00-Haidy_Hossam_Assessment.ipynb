```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless") 
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Initialize the Chrome driver
service = Service()
driver = webdriver.Chrome(service=service, options=chrome_options)

# Function to scrape one page from MobileMasr
def scrape_page(driver):
    data = []
    time.sleep(3)  # wait for page to fully load

    # find all product cards
    cards = driver.find_elements(By.CLASS_NAME, "product-card")
    print("Number of product cards found on this page:", len(cards))

    for i, card in enumerate(cards, 1):
        try:
            # product name
            try:
                name = card.find_element(By.CSS_SELECTOR, "div[class*='h-[47px]']").text.strip()
            except:
                name = None

            # product price
            try:
                price = card.find_element(By.CLASS_NAME, "text-green-700").text.strip()
            except:
                price = None

            # seller or title
            try:
                seller = card.find_element(By.CLASS_NAME, "text-xs").text.strip()
            except:
                seller = None

            # product link
            try:
                url = card.find_element(By.TAG_NAME, "a").get_attribute("href")
            except:
                url = None

            data.append({
                "Product Name": name,
                "Price": price,
                "Seller/Title": seller,
                "Listing URL": url,
                "Location": None
            })

        except Exception as e:
            print("Error while processing card", i, ":", e)
            continue

    print("Total products extracted from this page:", len(data))
    return data
```

```python
all_data = []
base_url = "https://mobilemasr.com/search?q=iphone13"

# open the first search page
driver.get(base_url)
time.sleep(5)

page_number = 1

while True:
    print("Scraping page", page_number, "...")

    # get products from current page
    page_data = scrape_page(driver)
    all_data.extend(page_data)

    # check if there is a next page
    try:
        next_button = driver.find_element(By.XPATH, "//button[@aria-label='التالي']")
        is_disabled = next_button.get_attribute("disabled")

        if is_disabled:
            print("Next button is disabled, last page reached.")
            break
        else:
            print("Going to the next page...")
            driver.execute_script("arguments[0].click();", next_button)
            page_number += 1
            time.sleep(5)  # wait for the next page to load

    except Exception as e:
        print("No more pages or an error occurred:", e)
        break

print("Scraping completed. Total products collected:", len(all_data))
```

```python
# convert all scraped data into a DataFrame
df = pd.DataFrame(all_data)

# save the data to a CSV file
output_file = "assessment_mobilemasr.csv"
df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(df.head())
```

```python
chrome_options = Options()

chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-blink-features=AutomationControlled")
chrome_options.add_argument("--start-maximized")

driver_dubizzle = webdriver.Chrome(options=chrome_options)
driver_dubizzle.get("https://dubai.dubizzle.com/classified/search/?keyword=samsung")

time.sleep(5)
print("Dubizzle page opened successfully")
```

```python
def scrape_dubizzle_page(driver_dubizzle):
    data_dubizzle = []

    # Wait a few seconds for the page to fully load
    time.sleep(3)

    # Find all product cards using the listing selector
    cards_dubizzle = driver_dubizzle.find_elements(By.CSS_SELECTOR, "a[data-testid^='listing-']")
    print("Found", len(cards_dubizzle), "product cards on this page.")

    # Loop through each product card and extract details
    for i, card in enumerate(cards_dubizzle, 1):
        try:
            # Extract product name
            try:
                name_dubizzle = card.find_element(By.CSS_SELECTOR, "h2[data-testid='subheading-text']").text.strip()
            except:
                name_dubizzle = None

            # Extract price
            try:
                price_dubizzle = card.find_element(By.CSS_SELECTOR, "div[data-testid='listing-price']").text.strip()
            except:
                price_dubizzle = None

            # Extract seller or title
            try:
                seller_title_dubizzle = card.find_element(By.CSS_SELECTOR, "h2[data-testid='subheading-text']").get_attribute("title")
            except:
                seller_title_dubizzle = None

            # Extract product URL
            try:
                url_dubizzle = card.get_attribute("href")
            except:
                url_dubizzle = None

            # Extract location if available
            try:
                location_dubizzle = card.find_element(By.CSS_SELECTOR, "h4.MuiTypography-body1").text.strip()
            except:
                location_dubizzle = None

            # Store all product details in a dictionary
            data_dubizzle.append({
                "Product Name": name_dubizzle,
                "Price": price_dubizzle,
                "Seller/Title": seller_title_dubizzle,
                "Listing URL": url_dubizzle,
                "Location": location_dubizzle
            })

        except Exception as e:
            # Handle unexpected errors for each card
            print("Error in card", i, ":", e)
            continue

    # Print total extracted products
    print("Extracted", len(data_dubizzle), "products successfully.")
    return data_dubizzle
```

```python
all_data_dubizzle = []
page_number_dubizzle = 1
max_pages = 5  # Stop after this many pages

print("\nStart paginated scraping on Dubizzle ...")

while True:
    print(f"\nScraping Dubizzle page {page_number_dubizzle} ...")

    # Scrape data from the current page
    page_data_dubizzle = scrape_dubizzle_page(driver_dubizzle)
    all_data_dubizzle.extend(page_data_dubizzle)
    print(f"Collected so far: {len(all_data_dubizzle)} listings total.")

    # Stop if page limit reached
    if page_number_dubizzle >= max_pages:
        print(f"Reached page limit ({max_pages}). Stopping pagination.")
        break

    # Try to find the "Next" button
    try:
        next_btn = driver_dubizzle.find_element(By.CSS_SELECTOR, "a[data-testid='page-next'], a.next_button")
    except Exception as e:
        print(f"Next button not found or not present: {e}")
        break

    # Check if the next button is disabled or has no link
    href = next_btn.get_attribute("href")
    aria_disabled = next_btn.get_attribute("aria-disabled")
    if (not href) or (aria_disabled and aria_disabled.lower() == "true"):
        print("Next button disabled or no href — last page reached.")
        break

    # Go to the next page
    print("Moving to next page ...")
    driver_dubizzle.execute_script("arguments[0].click();", next_btn)
    time.sleep(5)  # Give some time for the page to load
    page_number_dubizzle += 1

print(f"\nDubizzle pagination finished. Total listings collected: {len(all_data_dubizzle)}")

# Save Dubizzle results 
if all_data_dubizzle:
    df_dubizzle = pd.DataFrame(
        all_data_dubizzle,
        columns=["Product Name", "Price", "Seller/Title", "Listing URL", "Location"]
    )
    output_file_dubizzle = "assessment_dubizzle.csv"
    df_dubizzle.to_csv(output_file_dubizzle, index=False, encoding="utf-8-sig")
    print(f"\nDubizzle data saved to: {output_file_dubizzle}")
    print(df_dubizzle.head())
else:
    print("No Dubizzle data collected — nothing to save.")
```
